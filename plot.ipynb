# Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Plot test accuracy

```python
SMALL_SIZE = 16
MEDIUM_SIZE = 20
BIGGER_SIZE = 24

acc_dict = {
    'SAS': [], # Resnet50, Cifar100
    'Random': [58.65, 62.34, 66.3, 67.69],
    'MAE, IN1K': [59.15, 64.42, 66.27, 66.83],
    'Sup, IN1K': [59.66, 64.56, 66.7, 67.69],
    'DINO, IN1K': [60.61, 64.77, 66.86, 67.24],
    'Sup, IN21K': [59.02, 64.52, 66.57, 67.62]
}

X = np.array([20, 40, 60, 80])
acc_full = 69.13
list_method = ['Sup, IN1K', 'Sup, IN21K', 'MAE, IN1K', 'DINO, IN1K']
list_color = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown']

# Configure plot
plt.figure(figsize=(6.4,4.8))
plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=SMALL_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

ax = plt.subplot()

for idx, method in enumerate(list_method):
    list_acc = acc_dict[method]
    plt.plot(X, np.array(list_acc), lw=2, linestyle='-', marker='o', color=list_color[idx], label=method)

plt.axhline(y=acc_full, linestyle='-.', color='black', label='Full data')
plt.ylim(58, 70)
plt.grid()
plt.legend(loc="lower right")
plt.xlabel("Subset size (%)")
plt.ylabel("Top-1 Accuracy (%)")
# plt.title("Effect of proxy model (CIFAR100)")
plt.savefig('effect_of_proxy_model.pdf', bbox_inches='tight')
plt.show()
```
